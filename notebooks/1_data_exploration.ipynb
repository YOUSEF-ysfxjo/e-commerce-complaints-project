```python
import pandas as pd

file_path = '../data/reviews.csv'

with open(file_path, 'r', encoding='utf-8') as f:
    header = f.readline().strip().split(',')
print("Available columns:", header)

chunk_size = 100000
for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size)):
    print(f"\nProcessing chunk {i+1} (rows {i*chunk_size:,} to {(i+1)*chunk_size:,})")
    
    if 'rating' in chunk.columns:
        print(f"Average rating: {chunk['rating'].mean():.2f}")
    if 'app_id' in chunk.columns:
        print(f"Unique apps: {chunk['app_id'].nunique()}")
    
    print("\nFirst row of this chunk:")
    display(chunk.iloc[0])
    
    break
```

```python
import pandas as pd

file_path = '../data/reviews.csv'
output_path = '../data/processed/filtered_complaints.csv'
chunk_size = 100000

complaints_list = []

print("Starting the full data filtering process...")

for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size)):
    print(f"\n{'='*50}")
    print(f"Processing chunk {i+1}...")

    complaints_chunk = chunk[chunk['rating'] <= 2]
    complaints_list.append(complaints_chunk)

print("\nAll chunks processed. Now combining them into a final DataFrame...")

final_complaints_df = pd.concat(complaints_list, ignore_index=True)

final_complaints_df.to_csv(output_path, index=False)

print("-" * 50)
print("SUCCESS! ")
print(f"Total complaints found: {len(final_complaints_df):,}")
print(f"Filtered data has been saved to: {output_path}")
```
